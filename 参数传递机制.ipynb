```python
def foo(arg):
    arg +=1 
    
a = 2
foo(a)
print(a)
```

```python
def foo(arg):
    arg.append(6)
    
a = [1,2,3]
foo(a)
print(a)
```

整数型是不可变类型，而列表是可变类型，当整数型 arg += 1 的时候，会开辟一个新的空间来创造新的内存地址给引用，而列表 arg.append(6) 的时候，因为它是可变的，所以不会创建新的内存地址，只是在当前做改变，所以它们都指向了同一个内存地址。

当你的不可变类型变量希望通过函数进行“数据改变”的时候

```python
def foo(arg):
    return arg+1

a = 2
a = foo(a)
print(a)
```

```python
def foo(arg):
    arg +[4]
    
a = [1,2,3]
foo(a)
print(a)
```
